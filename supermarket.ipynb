```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
market=pd.read_csv('C:\\Users\\reine\\Desktop\\OTHER\\datasets\\supermarket.csv')
market.head(10)
```

```python
market.info()
```

```python
#we need the column "Date" to change it to a datetime datatype
market['Time']=pd.to_datetime(market['Time'])

#make a new column with just the hour 
market['Hour']=(market['Time']).dt.hour
market['Hour'].unique()
```

```python
market.describe()
```

```python
market['Hour'].unique()
```

```python
line=market.groupby('Hour').sum()
line.head()
```

```python
fig = px.line(line, x=line.index, y="Quantity",template='plotly_dark',title="Product Sales per Hour",color_discrete_sequence=['REd'])
fig.show()
```

```python
fig = px.histogram(market, x="Gender",template='plotly_dark',color_discrete_sequence=['Purple'],opacity=0.7,title='Gender Count')
fig.show()
```

```python
fig = px.box(market, y="Rating",x='Branch',template='plotly_dark',title='Ratings by Branch',color_discrete_sequence=['Purple'])
fig.show()
```

Product Analysis

```python
fig = px.histogram(market, x="Quantity",y='Product line',template='plotly_dark',opacity=0.7,color_discrete_sequence=['Purple'],title='Quantity per Product Line')
fig.show()
```

```python
fig = px.box(market, x="Total",y='Product line',template='plotly_dark',color_discrete_sequence=['Purple'],title='Total per Product Line')
fig.show()
```

```python
fig = px.strip(market, x='Total',y='Product line',template='plotly_dark',color='Gender',color_discrete_map={"Male": "#F3CCFF", "Female": "#570A57"},title='Total based on Gender')
fig.show()
```

```python
fig = px.box(market, x='gross income',y='Product line',template='plotly_dark',color_discrete_map={"Male": "#F3CCFF", "Female": "#570A57"},color_discrete_sequence=['Purple'],title='Gross Income per Product line')
fig.show()
```

```python
fig = px.box(market, x='Rating',y='Product line',template='plotly_dark',color_discrete_map={"Male": "#F3CCFF", "Female": "#570A57"},color_discrete_sequence=['Purple'],title='Ratings per Product Line')
fig.show()
```
